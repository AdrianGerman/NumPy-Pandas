```python
import numpy as np
```

```python
array = np.array([[1,2,3], [4,5,6]])
print(array.ndim)
print(array.shape)
print(array.dtype)
```

```python
z = np.array(3, dtype=np.uint8)
print(z)
```

```python
double_array = np.array([1,2,3], dtype='d')
print(double_array)
```

```python
z = z.astype(np.float64)
print(z)
```

```python
array = np.array([[1,2,3],[4,5,6]])
add = np.sum(array)
print(add)
```

```python
mean = np.mean(array)
print(mean)
```

```python
std = np.std(array)
print(std)
```
