# **Pivot Table y Reshape**

```python
import pandas as pd

file_path = "/content/Online_Retail.csv"
sales_data = pd.read_csv(file_path, encoding="windows-1252")
```

```python
pivot_table = pd.pivot_table(sales_data, values = "Quantity", index = "Country",
                             columns="StockCode", aggfunc="sum")
print(pivot_table)
```

```python
df = pd.DataFrame({
    "A": ["foo", "bar", "baz"],
    "B": [1,2,3],
    "C": [4,5,6]
})
print(df)
```

```python
df_stacked = df.stack()
print(df_stacked)
```

```python
df_unstacked = df_stacked.unstack()
print(df_unstacked)
```
