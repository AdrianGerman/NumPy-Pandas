```python
import pandas as pd

file_path = "/content/Online_Retail.csv"
sales_data = pd.read_csv(file_path, encoding="windows-1252")
```

```python
# crear dataframes de ejemplo
df1 = pd.DataFrame({
    "key": ["A", "B", "C"],
    "value1": [1,2,3]
})

df2 = pd.DataFrame({
    "key": ["B","C", "D"],
    "value2": [4,5,6]
})

print(df1)
print(df2)
```

```python
inner_merged = pd.merge(df1, df2, on="key", how="inner")
print(inner_merged)
```

```python
outer_merged = pd.merge(df1, df2, on="key", how="outer")
print(outer_merged)
```

```python
left_merged = pd.merge(df1,df2, on="key", how="left")
print(left_merged)
```

```python
right_merged = pd.merge(df1,df2, on="key", how="right")
print(right_merged)
```

```python
df3 = pd.DataFrame({
        'A': ['A0', 'A1', 'A2'],
        'B': ['B0', 'B1', 'B2']
    })
df4 = pd.DataFrame({
        'A': ['A3', 'A4', 'A5'],
        'B': ['B3', 'B4', 'B5']
    })
print(df3)
print(df4)
```

```python
vertical_concat = pd.concat([df3,df4])
print(vertical_concat)
```

```python
vertical_concat = pd.concat([df3,df4], axis=1)
print(vertical_concat)
```

```python
df5 = pd.DataFrame({
    'A': ['A0', 'A1', 'A2'],
    'B': ['B0', 'B1', 'B2']
}, index=['K0', 'K1', 'K2'])

df6 = pd.DataFrame({
    'C': ['C0', 'C1', 'C2'],
    'D': ['D0', 'D1', 'D2']
}, index=['K0', 'K2', 'K3'])

print(df5)
print(df6)
```

```python
joined = df5.join(df6, how="inner")
print(joined)
```
