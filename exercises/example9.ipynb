# **Cálculos matriciales**

```python
import numpy as np
```

```python
A = np.array([[1,2],[3,4]])
B = np.array([[5,6],[7,8]])

#SUMA
sum = A + B
print(A)
print(B)
print(sum)
```

```python
#MULTIPLICACIÓN
product = np.dot(A, B)
print(product)
```

```python
A = np.array([[1,2],[3,4]])
det_A = np.linalg.det(A)
inv_A = np.linalg.inv(A)
print(A)
print(det_A)
print(inv_A)
```

```python
#Resolver Ax = b
b = np.array([9,11])
x = np.linalg.solve(A,b)
print(x)
```
