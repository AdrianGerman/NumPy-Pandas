# **IndexaciÃ³n y Slicing**

```python
import numpy as np
```

```python
array = np.array([10,20,30,40,50])
print(array[1])
print(array[-1])
```

```python
print(array[1:4])
```

```python
array = np.array([10,20,30,40,50])
bool_index = array > 25
print(bool_index)
print(type(bool_index))
```

```python
index = [2,3,4]
print(array[index])
```

```python
array = np.random.randint(1,10, size=(3,3))
print(array)
print(array[0,1])
```

```python
print(array[:2,:2])
```
